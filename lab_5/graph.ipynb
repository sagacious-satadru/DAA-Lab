## Graph for plotting time taken 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data
# 500, 1000, 5000, 10000
x_random = [0.000031, 0.000059, 0.000317, 0.000620]
x_ascend = [0.000006, 0.000009, 0.000042, 0.000029]
x_descend = [0.000006, 0.000010, 0.000045, 0.000057]
y = [500, 1000, 5000, 10000]

# plot data
plt.plot(y, x_random, '-o', label='Random')
plt.plot(y, x_ascend, '-o', label='Ascend')
plt.plot(y, x_descend, '-o', label='Descend')
plt.legend()
plt.xlabel('Number of elements')
plt.ylabel('Time(s)')
plt.title('Time taken')
plt.show()
```
